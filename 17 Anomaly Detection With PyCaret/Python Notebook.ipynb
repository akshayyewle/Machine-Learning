```python
from pycaret.datasets import get_data
Anomaly_Data = get_data('anomaly',verbose=False)

print(type(Anomaly_Data))
print(Anomaly_Data.shape)
Anomaly_Data.head()
```

# **Anomaly Detection**

```python
from pycaret.anomaly import *
setup(data = Anomaly_Data)
```

# **Data Model**

## **Create Model**

* https://pycaret.readthedocs.io/en/stable/api/anomaly.html#pycaret.anomaly.predict_model

```python
Model01 = create_model('knn')
```

## **Plot Model**
* https://pycaret.readthedocs.io/en/stable/api/anomaly.html#pycaret.anomaly.plot_model

```python
plot_model(Model01)
```

## **Prediction**

* https://pycaret.readthedocs.io/en/stable/api/classification.html#pycaret.classification.predict_model

```python
predict_model(Model01,data = Anomaly_Data)
```

## **Assign Model**
* https://pycaret.readthedocs.io/en/stable/api/anomaly.html#pycaret.anomaly.assign_model

```python
Model01_Results = assign_model(Model01)
Model01_Results.head(10)
```

## **Save Model**

```python
save_model(Model01,'Model01.h5')
```

# **Multi-Model Comparison**

## **Create Classifier**

```python
from pycaret.classification import * 
Classifier01 = setup(data = Model01_Results,target = 'Anomaly')
```
