# **Setup**

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
```

# **Data Import**

```python
iris = load_iris(as_frame=True)
X_iris = iris.data[["petal length (cm)", "petal width (cm)"]].values
y_iris = iris.target
```

# **Data Model**

```python
Model01 = DecisionTreeClassifier(max_depth=2, random_state=42)
Model01.fit(X_iris, y_iris)
```

```python
Model01.tree_
```

## **Visualize Decision Tree**

```python
from sklearn.tree import export_graphviz

export_graphviz(Model01,out_file="iris_tree.dot",
                feature_names=["petal length (cm)", "petal width (cm)"],
                class_names=iris.target_names,
                rounded=True,filled=True)

from graphviz import Source
Source.from_file("iris_tree.dot")
```

## **Predict Class & Class Probability**

```python
# Predict Class Probability For Input Variables 5,1.5
Model01.predict_proba([[5, 1.5]]).round(3)
```

```python
# Predict Class For Input Variables 5,1.5
Model01.predict([[5, 1.5]]).round(3)
```
