## **Read All Files**

```python
len(df_route_cols)
```

```python
len(df_battery_cols)
```

```python
df = pd.DataFrame(columns=df_route_cols+df_battery_cols)
df.head()
```

```python
for file in os.listdir(Folder_Input):
# for file in ['TripA01.csv']:
    if file.endswith(".csv"):
        # clear_output(wait=True)
        print('Main Dataframe Shape: ',df.shape)
        battery_data = pd.read_csv(f'{Folder_Input}/{file}', sep=';',encoding='iso-8859-1')
        
        # Read Battery Data
        print(file,battery_data.shape)
        # display(battery_data.head())

        # Filter Route Data
        route_data = df_sample01[df_sample01['Trip']=='TripA01']
        # display(route_data.head())
        route_data_list = list(df_sample01[df_sample01['Trip']=='TripA01'].loc[0,:].values) 
        # print(route_data)

        # Concat Route & Battery Data
        route_battery_data = pd.DataFrame(columns=df_route_cols + df_battery_cols)
        print('Check 01:',route_battery_data.shape)
        route_battery_data = pd.concat([route_battery_data,battery_data],axis=0)
        print('Check 02:',route_battery_data.shape)
        route_battery_data.loc[:,df_route_cols] = route_data_list
        print('Check 03:',route_battery_data.shape)
        print('\n')
        # print(route_battery_data.shape)
        # display(route_battery_data.head())

        # Concatenate File Data To Main Dataframe
        df = pd.concat([df,route_battery_data],axis=0)
```
